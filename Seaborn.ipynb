```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")  # built-in dataset

sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Tip vs Total Bill (by Gender & Smoker)")
plt.show()
```

```python
tips = sns.load_dataset("tips")  # built-in dataset
#hue="sex" → different colors for male/female.
#style="smoker" → different markers for smoker/non-smoker.
sns.scatterplot(x="total_bill", y="tip", hue="sex", style="smoker", data=tips)
plt.title("Tip vs Total Bill (by Gender & Smoker)")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Average Total Bill by Day")
plt.show()
```

```python

sns.barplot(x="day", y="total_bill", data=tips, palette="coolwarm")
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.histplot(tips["tip"], bins=20, kde=True, color="violet")
plt.title("Distribution of Tips")
plt.show()
#kde=True adds smooth curve for probability distribution.
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set2")
plt.title("Total Bill Distribution by Day")
plt.show()
```

```python
sns.violinplot(x="sex", y="tip", data=tips, palette="pastel")

plt.show()
```

```python
sns.pairplot(tips, hue="sex")
plt.show()
```

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
