```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)       # line plot
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Plot")
plt.show()
```

```python
months=['jan','feb','mar','apr','may','jun']
sales=[15,67,23,5,12,78]
plt.bar(months,sales,color="red")
plt.xlabel("months")
plt.ylabel("sales")
plt.title("bar plot")
plt.show()
```

```python


months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun']
sales = [15, 67, 23, 5, 12, 78]

plt.pie(sales, labels=months, autopct='%1.2f%%')  # note double % to escape
plt.title("Monthly Sales Distribution")
plt.show()

```

```python

data = np.random.normal(size=1000)

plt.hist(data, bins=30)

plt.show()
```

```python
x = np.linspace(0, 10, 100)   # 100 points between 0 and 10
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with 2 subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Left plot (sine wave)
axs[0].plot(x, y1, color="blue")
axs[0].set_title("Sine Wave")

# Right plot (cosine wave)
axs[1].plot(x, y2, color="red")
axs[1].set_title("Cosine Wave")

plt.tight_layout()  # adjusts spacing
plt.show()
```

```python
# Random data
x = np.random.rand(50)   # 50 random points (0 to 1)
y = np.random.rand(50)
z = np.random.rand(50)

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, color="red", marker="o")

ax.set_title("3D Scatter Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

plt.show()
```
